```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,0:2],df.iloc[:,-1],test_size=0.2,random_state=0)
```

### 🔢 What is Ordinal Encoding?
**Ordinal encoding** converts categorical values into integers based on order. Useful when categories have a natural ranking.

```python
from sklearn.preprocessing import OrdinalEncoder
oe = OrdinalEncoder(categories=[['Poor','Average','Good'],['School','UG','PG']])
```

```python
oe.fit(x_train)
```

```python
x_train = oe.transform(x_train)
```

```python
x_train
```
