## Encoding Categorical Variables
Applying Ordinal Encoding to convert categories into numeric values.

```python
import pandas as pd
import numpy as np
from sklearn.impute import SimpleImputer
from sklearn.preprocessing import OneHotEncoder
# Use OrdinalEncoder to convert categorical features to numeric
from sklearn.preprocessing import OrdinalEncoder
```

```python
df=pd.read_csv("covid_toy.csv")
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,:5],df.iloc[:,-1],test_size=0.2,random_state=0)
```

```python
#(trf2,SimpleImputer(),["fever"]),
# Use OrdinalEncoder to convert categorical features to numeric
#(trf3,OrdinalEncoder(categories[["mild","strong"]]),["cough"])
#(trf4,OneHotEncoder(sparse_output=False,drop=first),["gender","city"])
```

```python
# Combine different preprocessing steps for different columns
from sklearn.compose import ColumnTransformer 
# Transform the data using the fitted transformer
transformer = ColumnTransformer(transformers=[
    ('tnf1',SimpleImputer(),['fever']),
# Use OrdinalEncoder to convert categorical features to numeric
    ('tnf2',OrdinalEncoder(categories=[['Mild','Strong']]),['cough']),
    ('tnf3',OneHotEncoder(sparse_output=False,drop='first'),['gender','city'])
],remainder='passthrough')
```

```python
# Fit the transformer and transform the data
transformer.fit_transform(x_train)
```

```python
# Transform the data using the fitted transformer
transformer.transform(x_test)
```
