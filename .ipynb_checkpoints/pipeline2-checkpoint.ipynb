## ColumnTransformer
Using ColumnTransformer to apply preprocessing to specific columns.

```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
# Combine different preprocessing steps for different columns
from sklearn.compose import ColumnTransformer 
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.preprocessing import MinMaxScaler
# Create a machine learning pipeline
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
df=pd.read_csv("train.csv")
```

```python
df.drop(columns=["PassengerId","Name","Ticket","Cabin",],inplace=True)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df.iloc[:,0:1],test_size=0.2,random_state=0)
```

```python
x_train.shape,y_train.shape
```

```python
x_train
```

## ColumnTransformer
Using ColumnTransformer to apply preprocessing to specific columns.

```python
# Combine different preprocessing steps for different columns
trf1=ColumnTransformer([
    ("impute",SimpleImputer(),[2]),
    ("impute2",SimpleImputer(strategy="most_frequent"),[6])],
    remainder="passthrough")
```

## ColumnTransformer
Using ColumnTransformer to apply preprocessing to specific columns.

```python
# Combine different preprocessing steps for different columns
trf2=ColumnTransformer([
    ('ohe',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])],remainder="passthrough")
```

## ColumnTransformer
Using ColumnTransformer to apply preprocessing to specific columns.

```python
# Combine different preprocessing steps for different columns
trf3=ColumnTransformer([
    ("scale",MinMaxScaler(),slice(0,10))])
    
```

```python
trf4=DecisionTreeClassifier()
```

## Building Pipelines
Combining preprocessing and modeling steps using sklearn Pipeline.

```python
# Create a machine learning pipeline
pipe=Pipeline([
    ("trf1",trf1),
    ("trf2",trf2),
    ("trf3",trf3),
    ("trf4",trf4)])
```

```python
pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
