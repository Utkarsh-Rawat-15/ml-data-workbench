## ðŸ“¦ Importing Required Libraries

```python
# Import seaborn for visualization and matplotlib for display
import seaborn as sns 
import pandas as pd
```

```python
# Execute the code below
df=pd.read_csv("train.csv")
df.head()
```

```python
# Execute the code below
sns.countplot(x="Embarked", data=df)

```

```python
# Execute the code below
df['Sex'].value_counts().plot(kind='pie',autopct='%.2f')
```

```python
# Import seaborn for visualization and matplotlib for display

import matplotlib.pyplot as plt
plt.hist(df['Age'],bins=5)
```

```python
# Execute the code below
sns.histplot(x="Age", data=df,kde=True)
```

## ðŸ“¦ Box Plot

### ðŸ“¦ What is a Box Plot?
A **box plot** is a graphical representation of data that shows the distribution using minimum, first quartile (Q1), median, third quartile (Q3), and maximum. It helps identify **outliers** and **skewness**.

```python
# Create a box plot to understand distribution and outliers
sns.boxplot(x="Age", data=df)
```

```python
# Create a box plot to understand distribution and outliers
sns.boxplot(x="Pclass",y="Age",hue="Sex",data=df)
```

```python
# Create a box plot to understand distribution and outliers
sns.boxplot(x="Sex",y="Age",hue="Survived",data=df)
```

```python
# Execute the code below
sns.distplot(df[df['Survived']==0]['Age'],hist=False)
sns.distplot(df[df['Survived']==1]['Age'],hist=False)
```

## ðŸ”¥ Heatmap

### ðŸ”¥ What is a Heatmap?
A **heatmap** is a graphical representation of data where values are represented with color intensity. It's often used to show **correlation matrices** in datasets to understand relationships between variables.

```python
# Create a heatmap to visualize correlations between numeric features

sns.heatmap(pd.crosstab(df['Pclass'],df['Survived']))
```

```python
# Execute the code below
sns.pairplot(df)
```

```python
# Execute the code below
sns.lineplot(x="Age",y="Fare",data=df)
```
