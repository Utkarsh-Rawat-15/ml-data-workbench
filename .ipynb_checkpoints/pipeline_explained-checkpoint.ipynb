### üîÉ What is Min-Max Scaling?
**MinMaxScaler** rescales data to a given range, typically [0, 1], preserving the shape of the original distribution.

```python
import pandas as pd 
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.impute import SimpleImputer
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import MinMaxScaler
from sklearn.feature_selection import SelectKBest,chi2
from sklearn.preprocessing import OneHotEncoder
from sklearn.tree import DecisionTreeClassifier
from sklearn.pipeline import Pipeline,make_pipeline
```

```python
df=pd.read_csv("train.csv")
```

```python
df.head()
```

```python
df.drop(columns=['PassengerId','Name','Ticket','Cabin'],inplace=True)
```

```python
df.head()
```

```python
x_train,x_test,y_train,y_test=train_test_split(df.iloc[:,1:],df['Survived'],test_size=0.2,random_state=0)
```

```python
x_train
```

### üß± What is a ColumnTransformer?
A **ColumnTransformer** allows you to apply different preprocessing pipelines to different columns of a dataset (e.g., scaling numeric columns and encoding categorical ones).

```python
trf1=ColumnTransformer([
    ("Impute_age",SimpleImputer(),[2]),
    ("impute_embarked",SimpleImputer(strategy='most_frequent'),[6])],
    remainder="passthrough")
```

### üß± What is a ColumnTransformer?
A **ColumnTransformer** allows you to apply different preprocessing pipelines to different columns of a dataset (e.g., scaling numeric columns and encoding categorical ones).

```python
trf2=ColumnTransformer([
    ('ohe_sex_embarked',OneHotEncoder(sparse_output=False,handle_unknown='ignore'),[1,6])],
    remainder="passthrough")
```

### üîÉ What is Min-Max Scaling?
**MinMaxScaler** rescales data to a given range, typically [0, 1], preserving the shape of the original distribution.

```python
trf3=ColumnTransformer([
    ("scale",MinMaxScaler(),slice(0,10))])
    
```

### üîç What is SelectKBest?
**SelectKBest** is a feature selection method that selects the top 'k' features based on a scoring function like `chi2` or `f_classif`.

```python
trf4 = SelectKBest(score_func=chi2,k=8)
```

```python
trf5=DecisionTreeClassifier()
```

### üß± What is a Pipeline?
A **Pipeline** in scikit-learn is a tool to chain preprocessing steps and models together. It helps in keeping the workflow clean, repeatable, and less error-prone.

```python
pipe=Pipeline([
    ("trf1",trf1),
    ("trf2",trf2),
    ("trf3",trf3),
    ("trf4",trf4),
    ("trf5",trf5)
])
```

```python
pipe.fit(x_train,y_train)
```

```python
from sklearn.metrics import accuracy_score
```

```python
y_pred=pipe.predict(x_test)
```

```python
accuracy_score(y_test,y_pred)
```
